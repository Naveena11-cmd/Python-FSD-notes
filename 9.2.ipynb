### MRO : Method Resolution Order 

```python
class A:
    def r(self):
        print("Class A")
class B(A):
    def r(self):
        print("Class B")
class C(A):
    def r(self):
        print("Class C")
class D(B,C):
    pass
obj=D()
obj.r()
print(D.mro())
```

## Abstract class 

```python
from abc import ABC, abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod    
    def draw(self):
        pass
class Circle(shape):
    def __init__(self):
        super().__init__("Circle")
    def draw(self):
        print("Drawing: ",self.shape_name)
```

```python
obj=Circle()
obj.draw()
```

```python
from abc import ABC, abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod    
    def draw(self):
        pass
class Circle(shape):
    def __init__(self):
        super().__init__("Circle")
    def draw_circle(self):
        print("Drawing: ",self.shape_name)
obj=Circle()
obj.draw_circle()
```

1.                                                                                                                              - Book : name, no of authors(n), authors(list), publisher, isbn, year
- CourseBook : name, no of authors(n), authors(list), publisher, isbn, year, course

```python
class Book:
    def __init__(self,name,n,authors,publisher,isbn,year):
        self.name=name
        self.n=n
        self.authors=authors
        self.publisher=publisher
        self.isbn=isbn
        self.year=year
    def display(self):
        print("Book name : ",self.name)
        print("Number of authors : ",self.n)
        print("List of authors : ",self.authors)
        print("Publisher : ",self.publisher)
        print("ISBN : ",self.isbn)
        print("Year : ",self.year)
class CourseBook(Book):
    def __init__(self,name,n,authors,publisher,isbn,year,course):
        super().__init__(name,n,authors,publisher,isbn,year)
        self.course=course
    def display(self):
        super().display()
        print("Course name : ",self.course)
```

```python
obj1=CourseBook("Harry Potter",1,["a"],"xyz",123, 2025,"fiction")
```

```python
obj1.display()
```

```python
class Student:
    def __init__(self,rn,name,age,total):
        self.rn=rn
        self.name=name
        self.age=age
        self.total=total
    def display(self):
        print("Name : ",self.name)
        print("Roll number : ",self.rn)
        print("Age : ",self.age)
        print("Total marks : ",self.total)
    def __eq__(self,other):
        if self.total==other.total:
            return True
        else:
            return False
```

```python
s1=Student(1,"Naveena",18,85)
s2=Student(2,"Aryan",20,80)
if s1==s2:
    print(f"{s1.name} and {s2.name} have same marks!")
else:
    print(f"{s1.name} and {s2.name} doesn't have same marks!")
```

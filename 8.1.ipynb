# Unit-8 : OOP(Abstraction & Encapsulation) and Exception Handling

1. try : block lets you test a block of code for errors
2. except : block lets you handle the error
3. else : block lets you execute the code when there is no error
4. finally : block lets you execute code regardless of the result of try and except

**Exception Handling**

```python
def div(a,b):
    return a/b
try:
    print(div(5,0))
except:
    print("Exception occured")
else:
    print("no exception")
finally:
    print("Done!")
```

**Custom Exception**

```python
# self = this of java
class Naveena(Exception):
    def __init__(self,arg):
        self.msg=arg
```

```python
def add(a,b):
    if type(a)==str:
        raise Naveena("Personalized error message")
    if type(b)==str:
        raise TypeError("b should be integer")
    return a+b
add("abc","def")
```

```python
balance=10000
def sub(balance,withdraw):
    amount=balance-withdraw
    print("amount = ",amount)
    try:
        if amount<=2000:
            raise Naveena("Balance is less than 2000 ")
    except Naveena as e:
        print(e)
withdraw=int(input("Enter amount you want to withdraw : "))
sub(balance,withdraw)
```

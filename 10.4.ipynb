```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(expense,income,color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(expense,income,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive')
plt.fill_between(time,income,expense,where=(income<expense),color='red',alpha=0.25,label='negative')
plt.xlabel("time")
plt.ylabel("income/expense comparision")
plt.title("Area fill between two lines")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(expense,income,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<expense),color='red',alpha=0.25,label='negative',interpolate=True)
plt.xlabel("time")
plt.ylabel("income/expense comparision")
plt.title("Area fill between two lines")
plt.legend()
plt.show()
```

### Polymorphism : overloading and overriding                                                                                                               
- can write methods that has same name but different functionalities
- overloading : methods having same name but different arguments (compile time)
- overriding : child class methods having same name,parameters and return type that overrides the parent class method in derived class (run time)

```python
def sum_nos(*args):
    result=0
    for num in args:
        result+=num
    print("sum = ",result)
print("Similar to method overloading concept")
print("Single argument : ",end=" ")
sum_nos(10)
print("Two argument : ",end=" ")
sum_nos(30,2)
print("Multiple arguments : ",end=" ")
sum_nos(1,2,3)
```

```python
# compile time polymorphism error
def fun(a):
    print("x")
def fun(a,b):
    print("y")
fun(5)
```

```python
class A:
    def first(self):
        print("first function of class A")
    def second(self):
        print("second function of class A")

class B(A):
    def first(self):
        print("redefined function of class A in class B")
    def display(self):
        print("Display function of child class")
```

```python
obj=B()
obj.first()
```

```python
obj.second()
```

### Operator Overloading :                                                                                                                                                      
- giving extended meaning beyond their predefined operational meaning
- '+' operator : 
           1. Adds 2 intergers
           2. joins 2 strings
           3. merges 2 lists

```python
class A:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
obj1=A(1)
obj2=A(2)
obj3=A("Hello")
obj4=A("hi")
```

```python
print(obj1+obj2)
```

```python
print(obj3+obj4)
```

```python
print(A.__add__(obj1,obj2))
print(A.__add__(obj3,obj4))
```

```python
print(obj1.__add__(obj2))
print(obj3.__add__(obj4))
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            return True
        else:
            return False
obj1=A(5)
obj2=A(3)
if obj1>obj2:
    print("obj1 is greater")
else:
    print("obj2 is greater")
```

```python
1.  '+': __add__(self,other)
2.  '-' : __sub__(self,other)
3.  '*' : __mul__(self,other)
4.  '/' : __truediv__(self,other)
5.  '//' : __floordiv__(self,other)
6.  '%' : __mod__(self,other)
7.  '**' : __pow__(self,other)
8.  '>' : __gt__(self,other)
9.  '<' : __lt__(self,other)
10. '<=' : __le__(self,other)
11.  '>=' : __ge__(self,other)
12.  '==' : __eq__(self,other)
13.  '!=' : __ne__(self,other)
```

### Inheritance :                                                                                                                                                                   
- use of existing class to create a new class
- increases reusability of code
- with this, child class gains access to all the data memebers, functions and properties defined in parent class
- child class may also offer its particular functionalities to parent class

```python
# 1. SINGLE INHERITANCE
class parent:
    def f1(self):
        print("Hello parent")
class child(parent):
    def f2(self):
        print("Hello child")
obj=child()
obj.f1()
obj.f2()
```

```python
obj=child()
obj.f1()
obj.f2()
obj.f3()
```

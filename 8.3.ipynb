**Abstraction**

- Abstraction shows us functionalities anything holes, hiding all implementations or inner details
- Main goal of abstraction is to hide background details or any unnecessary implementation about the data so the users only see the required info it helps handling the complexity
- It can be achieved by using **abstract class**
- A class contains one or more abstract methods is called abstract class
- Abstract method do not contain any imp implementation of their own
- Abstract class can act like blue print of other classes

**Encapsulation**

- Process of binding data and corresponding methods together into a single unit
- Programming technique that binds the class members together and prevents from being accessed by other classes
- hides data from the access of outsiders


1. **public** : can be freely used inside or outside the class
2. **protected** : should not be used outside a class except inside subclass
3. **private** : inaccessible and invisible, neither read or write to these attribute except inside class

```python
class student:
    branch='CSE'
    def __init__(self,rollno,name):
        self.rollno=rollno
        self.name=name
    def enter_marks(self):
        self.marks=[]
        for i in range(5):
            self.marks.append(int(input("Enter marks of subject {i+1} : ")))
    def display(self):
        print("Name : ",self.name)
        print("Branch : ",self.branch)
        print("Rollno : ",self.rollno)
        
        if len(self.marks)>0:
            print("Marks : ",self.marks)
    def calculate_avg(self):
        sum=0
        for i in self.marks:
            sum+=i
        print("Avg marks : ",sum/5)
```

wap to create a class string and make methods to count uppercase, lowercase, vowels and consonants and spaces

```python
class String:
    def __init__(self):
        self.uppercase=0
        self.lowercase=0
        self.vowels=0
        self.consonants=0
        self.spaces=0
        self.s=input("Enter a string : ")
    def count_uppercase(self):
        for letter in self.s:
            if letter.isupper():
                self.uppercase+=1
    def count_lowercase(self):
        for letter in self.s:
            if letter.islower():
                self.lowercase+=1
    def count_vowels(self):
        for letter in self.s:
            if letter in "aeiouAEIOU":
                self.vowels+=1
    def count_consonants(self):
        for letter in self.s:
            if letter.isalpha() and (letter not in "aeiouAEIOU"):
                self.consonants+=1
    def count_spaces(self):
        for letter in self.s:
            if letter.isspace():
                self.spaces+=1
    def display(self):
        self.count_uppercase()
        self.count_lowercase()
        self.count_vowels()
        self.count_consonants()
        self.count_spaces()
        print("uppercase : ",self.uppercase)
        print("lowercase : ",self.lowercase)
        print("vowels : ",self.vowels)
        print("consonants : ",self.consonants)
        print("spaces : ",self.spaces)
```

```python
# enter pin, create pin, check balance, withdrawal
class bank:
    def __init__(self):
        self.pin=""
        self.bal=0
        self.menu()
    def menu(self):
        m=input(""""Menu : 
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw""")
        
        if m=='1':
            self.createPin()
        elif m=='2':
            self.changePin()
        elif m=='3':
            self.checkBal()
        elif m=='4':
            self.withdraw()
        else:
            exit()
        
    def createPin(self):
        newPin=input("Enter new pin : ")
        self.pin=newPin
        newBal=int(input("Enter initial balance : "))
        self.bal=newBal
        print("Pin created successfully!")
        self.menu()
        
    def changePin(self):
        oldPin=input("Enter old pin : ")
        if oldPin==self.pin:
            nPin=input("Enter your new pin : ")
            self.pin=nPin
            print("Pin changed successfully!")
        else:
            print("Enter valid pin : ")
        self.menu()
        
    def checkBal(self):
        self.menu()
        
    def withdraw(self):
        self.menu()
```

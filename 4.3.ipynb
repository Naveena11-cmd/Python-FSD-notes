# Tuple


- Immutable : Once we create tuple object, we cannot perform any changes in that object
- Tuple is read only version of list
- If our data is fixed and never changes, then we should go for tuple
- Insertion order is preserved
- Duplicates are allowed
- heterogenous objects are allowed
- We can differentiate duplicate object by using index 
- Tuple supports both positive and negative index
- We can represnt tuple elements with paranthesis and with commas
- paranthesis are optional but recommended 

```python
t=(10,)
print(type(t))
```

```python
t=10,20,30,401
print(type(t))
print(t)
```

```python
t=tuple(range(10,20,2))
print(t)
```

```python
t=(10,20,30)
print(t[2])
# print(t[4])
print(t[-2])
# print(t[50])
print(t[2:5])
print(t[2:100])
print(t[2:])
print(t[::2])
```

```python
# concatanation operator (+)
t1=(10,20,30)
t2=(30,40,50)
t=t1+t2
print(t)
```

```python
# Repetition operator(*)
t1=10,20,30
t2=t1*3
print(t2)
```

```python
# len
t=10,20,30,60
print(len(t))
```

```python
# count
t=10,20,49,49,29,87
print(t.count(49))
```

```python
# index
t=10,20,10,30,40
print(t.index(10))
print(t.index(50))
```

```python
s="abcdabcabc"
print(s.index("a"))
print(s.index("x"))
```

```python
# sorted
t=40,10,30,20
t1=sorted(t)
print(t1)
print(type(t1))
```

```python
s="xpyaz"
s1=sorted(s)
print(s1)
print(type(s1))
```

```python
t=40,10,30,20
t1=sorted(t,reverse=True)
print(t1)
```

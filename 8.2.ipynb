**Class :** 
- User defined data structure that binds the data members and methods into a single unit . 
- Class is a blue print or code template for object creation 
- Using a class, you can create as many objects you want

**Object :**
- Object is an instance of a class
- It is a collection of attributes(variables) and methods
- We use the object of a class to perform actions
1.  **States** : represents state of an object
2. **Behaviour** : obj method that represents behaviour

**Advantages :**
1. Flexibility of having multiple formd (POLYMORPHISM)
2. Reduce high complexity of code using ABSTRACTION
3. ENCAPSULATION : security
4. Reuse the code by child class (INHERITANCE)

```python
class emp():
    def __init__(self,name,age,salary):
        print("Address of self : ",id(self))
        self.name=name
        self.age=age
        self.salary=salary
    def display(self):
        print("Employee Name : ",self.name)
        print("Employee Age : ",self.age)
        print("Employee Salary : ",self.salary)
```

```python
emp1=emp("naveena", 18,30000)
print(id(emp1))
print(emp1.name)
# print(self.name)  NameError
emp1.display()
```

```python
class person:
    def __init__(self):
        self.name="naveena"
        self.age=18
    def update(self):
        self.age=20
    def compare(self,other):
        if self.age==other.age:
            print("Age is same")
        else:
            print("Age is different")
```

```python
p1=person()
p2=person()
print(p1.age,p2.age)
p1.compare(p2)
```

```python
p1=person()
p1.update()
p2=person()
print(p1.age,p2.age)
p1.compare(p2)
```

```python
class emp():
    def __init__(self,name,age,salary):
        self.name=name
        self.age=age
        self.salary=salary
    def __del__(self):
        print("object deleted")
    def display(self):
        print("Employee Name : ",self.name)
        print("Employee Age : ",self.age)
        print("Employee Salary : ",self.salary)
```

```python
emp1=emp("naveena", 18,30000)
emp2=emp("Ria",20,45000)
emp1.display()
emp2.display()
```

**To Delete :**

```python
emp1=emp("naveena", 18,30000)
emp2=emp("Ria",20,45000)
emp1.display()
emp2.display()
del emp1
emp1.display()
```

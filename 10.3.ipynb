```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,6])
y=np.array([0,10])
plt.plot(x,y)
plt.show()
```

```python
x=np.array([0,6])
y=np.array([0,10])
plt.plot(x,y,'o')  #circle marker
plt.show()
```

```python
x=np.array([0,6])
y=np.array([0,8])
plt.plot(x,y,'*') #star marker
plt.show()
```

```python
x=np.array([0,6])
y=np.array([0,10])
plt.plot(x,y,'v') #triangle
plt.show()
```

```python
x=np.array([0,6])
y=np.array([0,10])
plt.plot(x,y,'s') #square
plt.show()
```

=============    ===============================
character        description
=============    ===============================
``'.'``          point marker
``','``          pixel marker
``'o'``          circle marker
``'v'``          triangle_down marker
``'^'``          triangle_up marker
``'<'``          triangle_left marker
``'>'``          triangle_right marker
``'1'``          tri_down marker
``'2'``          tri_up marker
``'3'``          tri_left marker
``'4'``          tri_right marker
``'s'``          square marker
``'p'``          pentagon marker
``'*'``          star marker
``'h'``          hexagon1 marker
``'H'``          hexagon2 marker
``'+'``          plus marker
``'x'``          x marker
``'D'``          diamond marker
``'d'``          thin_diamond marker
``'|'``          vline marker
``'_'``          hline marker
=============    ===============================

**Line Styles**

=============    ===============================
character        description
=============    ===============================
``'-'``          solid line style
``'--'``         dashed line style
``'-.'``         dash-dot line style
``':'``          dotted line style
=============    ===============================

Example format strings::

    'b'    # blue markers with default shape
    'or'   # red circles
    '-g'   # green solid line
    '--'   # dashed line with default color
    '^k:'  # black triangle_up markers connected by a dotted line

**Colors**

The supported color abbreviations are the single letter codes

=============    ===============================
character        color
=============    ===============================
``'b'``          blue
``'g'``          green
``'r'``          red
``'c'``          cyan
``'m'``          magenta
``'y'``          yellow
``'k'``          black
``'w'``          white
=============    ===============================

and the ``'CN'`` colors that index into the default property cycle.

```python
x=np.array([1,2,6,8])
y=np.array([3,7,2,5])
plt.plot(x,y,marker='o')
plt.show()
```

```python
x=np.array([1,2,6,8])
y=np.array([3,7,2,5])
plt.plot(x,y,'*:c')
plt.show()
```

```python
x=np.array([1,2,6,8])
y=np.array([3,7,2,5])
plt.plot(x,y,'*-.m',ms=18,mec='b',mfc='k')
plt.show()
```

```python
y=np.array([3,7,2,5]) # default x value : 0,1,2,3
plt.plot(y,marker='o')
plt.show()
```

```python
y=np.array([3,7,2,5])
plt.plot(y,linestyle="dotted")
plt.show()
```

```python
y=np.array([3,7,2,5])
plt.plot(y,ls=":")
plt.show()
```

```python
y=np.array([3,7,2,5])
plt.plot(y,lw=5) #line width
plt.show()
```

**314** Write a Python program to create a Caesar encryption.                                                             
Note: In cryptography, a Caesar cipher, also known as Caesar's cipher, the shift cipher, Caesar's code or Caesar           
shift, is one of the simplest and most widely known encryption techniques. It is a type of substitution cipher in            
which each letter in the plaintext is replaced by a letter some fixed number of positions down the alphabet. For           
example, with a right shift of 3, A would be replaced by D, E would become H, and so on. The method is named             
after Julius Caesar, who used it in his private correspondence.                                                          
For Example:                                                                                                    
Input Text : LJIET ENG                                                                                                  
Shift : 3                                                                                                            
Cipher: OMLHW HQJ                                                                                                       

```python
s=input("Enter string : ")
shift=int(input("Enter a shift : "))
c=""
alpha="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in s:
    if i in alpha:
        i_index=(alpha.find(i)+shift)%len(alpha)
        c=c+alpha[i_index]
    else:
        c=c+i
print(c)
```

**396**The Syracuse (also called Collatz or Hailstone) sequence is generated by starting with a natural number and repeatedly
applying the following function until reaching 1 :                                                                
syr(x) = 1/2 if x is even; and                                                                                    
syr(x) = 3x + 1 if x is odd                                                                                        
For example, the Syracuse sequence starting with 5 is: 5, 16, 8, 4, 2, 1.                                               
Use of inbuilt function or math library is not allowed.                                                                    
This sequence will always go to 1 for every possible starting value.                                                       
Write a program that                                                                                                         
1.Gets a string value from the user                                                                                        
2.Prints the Syracuse sequence for that string value

```python
n=input("enter a number : ")
if n%2==0:
    even(n)
def even(n):
    while n!=1:
        n=n/2
        print(n)
def odd(n):
    while n!=1:
        n=3*n+1
        print(n)
else:
    odd(n)
```

**=====MODULES=====**

```python
%%writefile calculator.py
def add(a,b):
    print("sum : ",a+b)
def sub(a,b):
    print("difference : ",a-b)
def mul(a,b):
    print("multiplication : ",a*b)
def div(a,b):
    print("division : ",a/b)
def power(a,b):
    print("power : ",a**b)
def modulo(a,b):
    print("modulo : ",a%b)
def floor(a,b):
    print("floor : ",a//b)
```

```python
%%writefile myModule.py
def myString(s):
    print("Length : ",len(s))
    print("Upper : ",s.upper())
    print("Lower : ",s.lower())
    print("Capitalize : ",s.capitalize())
    print("Is upper : ",s.isupper())
    print("Is Lower : ",s.islower())
    print("Title : ",s.istitle())
    print("Alpha : ",s.isalpha())
    print("Digit : ",s.isdigit())
    print("Alnum : ",s.isalnum())
    
def myList(s):
    print("Length : ",len(s))
    s.append(100)
    print("append : ",s)
    s.extend([10,20,30])
    print("Extend : ",s)
    s.sort()
    print("Sort : ",s)
    s.reverse()
    print("Reverse : ",s)
```

```python
import myModule as m
m.myString("Naveena11")
```

```python
import myModule as m
l=[1,2,3,4,5]
m.myList(l)
```

**Date and Time module**
- date : year,month,day (YYYY-MM-DD)
- time : hour, minutes, seconds
- datetime : combines both

```python
from datetime import date
today=date.today()
print(today)
```

```python
from datetime import date
d=date(2025,13,30)
print(d)
```

```python
from datetime import date
d=date(2025,11,31)
print(d)
```

```python
from datetime import time
d=time(10,15,30)
print(d)
```

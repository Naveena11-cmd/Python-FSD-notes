**Changing the case of string**

```python
a="How are YoU"
print(a.upper())
print(a.lower())
# swapcase : changes upper to lower and lower to upper
print(a.swapcase())
# title : changes first letter of evry word in sentence to uppercase and rest to lower case
print(a.title())
# capitalize : only changes first letter of first word of string to uppercase and rest of the string to lowercase
print(a.capitalize())
```

**To check the types of characters present in the string** (answers only in True or False)                                 
1. isalnum() :(a to z , A to Z, 0 to 9) returns true if all characters are AlphaNumeric
2. isalpha(): (a to z , A to Z)
3. isdigit() : (0 to 9) or isnumeric() : (0 to 9 , roman numbers, fraction)
4. islower(): returns True when the whole string is in lower case (doesn't checks num or spl char)
5. isupper(): returns True when the whole string is in upper case (doesn't checks num or spl char)
6. istitle(): returns True only when the first digit of evry word in the string is capital(doesn't checks num or spl char)
7. isidentifier() 
8. isspace()

```python
t="company123"
print(t.isalnum())
print(t.isalpha())
print(t.isdigit())
print(t.islower())
print(t.isupper())
```

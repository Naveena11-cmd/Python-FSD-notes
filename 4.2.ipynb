**5. translate() with maketrans() function**                                                                        
1. maketrans() :                                                                                                         
- make translation table
- mapping of character to their replacement or to none for deletion
- Syntax : maketrans(from_char, to_char, delete_char)
- from_char and to_char length should be same
2. translate()
- applies to transition table created by maketrans()
- it returns new string with character replaced or deleted acc to the table

```python
import string
print(string.punctuation)
```

```python
print(len(string.punctuation))
```

```python
c="Py$@t/*hon"
t=c.maketrans("","",string.punctuation)
print(t)
```

```python
d=c.translate(t)
print(d)
```

```python
c="Py$@t/*hon!!"
t=c.maketrans("","","$@/*")
print(t)
d=c.translate(t)
print(d)
```

```python
c="Py$@t/*hon!!"
t=c.maketrans("n","m","$@/*")
print(t)
d=c.translate(t)
print(d)
```

```python
c="Py$@t/*hon!!"
t=c.maketrans("n","m",)
print(t)
d=c.translate(t)
print(d)
```

```python
# to change multiple characters
s="Hi Sam!!"
x="Sam"
y="Joe"
mytable=s.maketrans(x,y)
print(s.translate(mytable))
```

```python
s="Py$@*/th45729on"
countUpper=0
countLower=0
countAlpha=0
countDigit=0
countSplCh=0
sum=0
for i in s:
    if i.isalpha():
        countAlpha+=1
        if i.isupper():
            countUpper+=1
        else:
            countLower+=1
    elif i.isdigit():
        countDigit+=1
        sum=int(i)
    else:
        countSplCh+=1
print("upper count",countUpper)
print("lower count",countLower)
print("alphabetes count",countAlpha)
print("digits count",countDigit)
print("special character count",countSplCh)
print("sum of digits",sum)
print("avg",sum/countDigit)
```

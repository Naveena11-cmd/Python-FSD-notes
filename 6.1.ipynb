**File Access Modes**                                                                                           
- how file will be used once it is opened
- File Handle is like a cursor, which defines from where the data has to be read or written in the files
- modes also define the location of the File Handle in file
- there are 6 access modes (r,r+,w,w+,a,a+)   (r=read, a=append, w=write)

1. read only ('r'):
- open text for reading
- the handle is positioned at the begining of the file
- if file does not exist, it raises an error
- This is default mode in which file is opened
- Syntax : f=open("t.txt",'r') or f_open("t.txt")

2. read and write('r+') 
- open text for reading and writing
- the handle is positioned at the begining of the file
- if file does not exist, it raises an error
- Syntax : f=open("t.txt",'r+')

3. write only ('w'):
- open text for writing
- the handle is positioned at the begining of the file
- For existing file, data is truncated and over written (exisitng data gets deleted and new one is added)
- creates the file if doesn't exists already
- Syntax : f=open("t.txt",'w')

4. write and read ('w+'):
- open text for writing
- the handle is positioned at the begining of the file
- For existing file, data is truncated and over written (exisitng data gets deleted and new one is added)
- creates the file if doesn't exists already
- Syntax : f=open("t.txt",'w+')

5. append mode ('a'):
- open text for writing
- the handle is positioned at the end of the file
- For existing file, new data will be inserted at the end after exisitng data
- creates the file if doesn't exists already
- Syntax : f=open("t.txt",'a')

6. append mode ('a+'):
- open text for writing and read
- the handle is positioned at the end of the file
- For existing file, new data will be inserted at the end after exisitng data
- creates the file if doesn't exists already
- Syntax : f=open("t.txt",'a+')

```python
f=open("n.txt",'w')
l=[]
for i in range(5):
    name=input("Enter name : ")
    l.append(name+'\n')
f.writelines(l)
f.close()
```

wap to copy odd lines from one file to another

```python
f = open("f.txt")
l = f.readlines()[0::2]
f=open("odd.txt",'w')
for i in l:
    f.write(i)
f.close()
```

wap to replace a word in a file

```python
c=input("enter the word you want to change : ")
n=input("Enter the new word : ")
f=open("f.txt",'r')
l=f.read()
r=l.replace(c,n)
f.close()
f1=open("f.txt",'w')
f1.write(r)
f1.close()
```

wap to count the number of lines, words, characters and spaces from a text file

```python
line=0
word=0
char=0
space=0
with open("f.txt") as f:
    line=len(f.readlines())
    f.seek(0)
    x=f.read()
    word=len(x.split())
    char=len(x)-x.count(" ")-x.count('\n')
    space=x.count(" ")
    print("lines : ",line)
    print("words : ",word)
    print("characters : ",char)
    print("spaces : ",space)
```

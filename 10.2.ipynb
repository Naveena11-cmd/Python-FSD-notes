### 1. concatenate()

```python
import numpy as np
a1=np.array([1,2,3])
a2=np.array([4,5,6])
a3=np.array([7,8,9])
a=np.concatenate((a1,a2,a3))
print(a)
```

```python
import numpy as np
a1=np.array([[1,2],[3,4]])
a2=np.array([[5,6],[7,8]])
a=np.concatenate((a1,a2))
print(a)
```

```python
import numpy as np
a1=np.array([[1,2],[3,4]])
a2=np.array([[5,6],[7,8]])
a=np.concatenate((a1,a2),axis=0)
print(a)
```

```python
import numpy as np
a1=np.array([[1,2],[3,4]])
a2=np.array([[5,6],[7,8]])
a=np.concatenate((a1,a2),axis=1)
print(a)
```

### 3. where() : finds index

```python
import numpy as np
a1=np.array([1,2,3,4,5,4,6,4,3])
x=np.where(a1==4)
print(x)
```

```python
import numpy as np
a1=np.array([[1,2],[3,4],[5,4],[6,4]])
x=np.where(a1==4)
print(x)
```

```python
import numpy as np
a1=np.array([[[1,2,3],[4,5,6]],[[4,8,9],[10,4,12]]])
x=np.where(a1==4)
print(x)
```

```python
import numpy as np
a1=np.array([[1,2],[3,4],[5,4],[6,4]])
x=np.where(a1%2==0)
print(x)
```

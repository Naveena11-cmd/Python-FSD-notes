**A Disarium number is a number where the sum of its digits, each raised to the power of its respective position, equals the original number. For example, 175 is a Disarium number because \(1^{1}+7^{2}+5^{3}=1+49+125=175\)**

```python
def is_disarium(num):
    """
    Checks if a given number is a Disarium number.

    Args:
        num: The integer to check.

    Returns:
        True if the number is a Disarium number, False otherwise.
    """
    num_str = str(num)
    length = len(num_str)
    sum_digits_powered = 0

    for index, digit_char in enumerate(num_str):
        digit = int(digit_char)
        position = index + 1  # Positions are 1-indexed
        sum_digits_powered += digit ** position

    return sum_digits_powered == num

# Example usage:
number = int(input("Enter a number: "))

if is_disarium(number):
    print(f"{number} is a Disarium number.")
else:
    print(f"{number} is not a Disarium number.")
```

**wap to count total odd and even number and also do the sum of them**

```python
n=int(input("How many numbers you want to enter : "))
oddcount=0
oddsum=0
evencount=0
evensum=0
for i in range(n):
    numbers=int(input("enter the number : "))
    if numbers%2==0:
        evencount=evencount+1
        evensum+=numbers
    else:
        oddcount=oddcount+1
        oddsum+=numbers
print()
print("total even numbers : ",evencount)
print("Sum of even nmbers : ",evensum)
print("total odd numbers : ",oddcount)
print("Sum of odd nmbers : ",oddsum)
```

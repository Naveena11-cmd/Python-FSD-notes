# Tuple


- Immutable : Once we create tuple object, we cannot perform any changes in that object
- Tuple is read only version of list
- If our data is fixed and never changes, then we should go for tuple
- Insertion order is preserved
- Duplicates are allowed
- heterogenous objects are allowed
- We can differentiate duplicate object by using index 
- Tuple supports both positive and negative index
- We can represnt tuple elements with paranthesis and with commas
- paranthesis are optional but recommended 

```python
t=(10,)
print(type(t))
```

```python
t=10,20,30,401
print(type(t))
print(t)
```

```python
t=tuple(range(10,20,2))
print(t)
```

```python
t=(10,20,30)
print(t[2])
# print(t[4])
print(t[-2])
# print(t[50])
print(t[2:5])
print(t[2:100])
print(t[2:])
print(t[::2])
```

```python
# concatanation operator (+)
t1=(10,20,30)
t2=(30,40,50)
t=t1+t2
print(t)
```

```python
# Repetition operator(*)
t1=10,20,30
t2=t1*3
print(t2)
```

```python
# len
t=10,20,30,60
print(len(t))
```

```python
# count
t=10,20,49,49,29,87
print(t.count(49))
```

```python
# index
t=10,20,10,30,40
print(t.index(10))
print(t.index(50))
```

```python
s="abcdabcabc"
print(s.index("a"))
print(s.index("x"))
```

**tuple packing**

```python
a=10
b=20
c=30
d=40
t=a,b,c,d
print(t)
```

**tuple unpacking**

```python
t=(10,20,30,40)
a,b,c,d=t
print(a,b,c,d)
```

```python
# loop through tuple
t=("apple","banana","cherry")
for i in t:
    print(i)
```

```python
# loop through index
t=("apple","banana","cherry")
for i in range(len(t)):
    print(t[i])
```

```python
# using while loop
t=("apple","banana","cherry")
i=0
while i<len(t):
    print(t[i])
    i=i+1
```

```python
# reversed()
s="python"
print(list(reversed(s)))

t=(1,4,6,2,5,3)
print(list(reversed(t)))

r=range(5,9)
print(list(reversed(r)))

l=[1,2,3,4]
print(list(reversed(l)))
```

**enumerate()** : if you pass a string to enumerate(), the output will show index and value for each character of the string   
Syntax : enumerate(iterable,start=0)

```python
s="LJIET"
print(list(enumerate(s)))
l=["eat","sleep","repeat"]
print(list(enumerate(l)))
```

```python
s="LJIET"
print(list(enumerate(s,10)))
```

```python
s="LJIET"
obj1=list(enumerate(s))
for i in obj1:
    print(i)
```

```python
s="LJIET"
obj1=list(enumerate(s))
for i,j in obj1:
    print(i,'-',j)
```

```python
# print min and max even and odd
t=(1,3,4,6,9,12,13,14)
even=0
odd=0
evenC=0
oddC=0
for i in t:
    if i%2==0:
        even=even+i
        evenC=evenC+1
    else:
        odd=odd+i
        oddC=oddC+1
print("even count = ",evenC)
print("even sum = ",even)
print("odd count = ",oddC)
print("odd sum = ",odd)
```

1. wap to perform a circular shift on a list to the right direction

```python
n=[1,2,3,4,5,6,7]
s=int(input("Enter number of shifts : "))
s=s%len(n)
x=n[-s:]+n[:-s]
print(x)
```

```python
n=[1,2,3,4,5,6,7]
s=int(input("Enter number of shifts : "))
s=s%len(n)
for i in range(s):
    p=n.pop()
    n.insert(0,p)
print(n)
```

2. wap to print elements with frequency greater then a given value k

```python
l=[1,1,1,1,2,2,3,3,3,4,5,6,6,6]
k=int(input("Enter frequency : "))
m=[]
for i in range(len(l)):
    if l.count(i)>k:
        m.append(i)
print(m)
```

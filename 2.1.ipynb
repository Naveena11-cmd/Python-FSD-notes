**nested loop**

```python
for i in range(2):
    for j in range(2):
        print("i = ",i," j= ",j)
```

**break**

```python
for i in range(10):
    if i==7:
        print("stop")
        break
    print(i)
```

**continue**

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

**pass statement**

```python
for i in range(20):
    if i%9==0:
        print(i)
else:
            pass
```

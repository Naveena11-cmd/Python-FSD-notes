**wap to count the frequency of words from the given text file**

```python
f=open("p1.txt")
r=f.read().split().lower()
d={}
for i in r:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
for i in d:
    print(i,"-",d[i])
f.close()
```

**wap to accept string/lines from user till the user enters 'end'. each string entered by user should be a new line in file, save all the lines in text file and display only those lines which begines with capital letter**

```python
f=open("p4.txt",'w+')
while True:
    lines=input("Enter strings : ")
    if lines.lower() == 'end':
        break
    else:
        f.write(lines+'\n')
f.seek(0)
for line in f:
        if line.strip() and line.strip()[0].isupper():
            print(line[:-1])
```

**wap to read a text file and print words staring with i in reverse order**

```python
f=open("p4.txt")
r=f.read().split()
new=""
for i in r:
    if i[0]=='I' or i[0]=='i':
        new+=i[::-1]+" "
    else:
        new+=i+" "
print(new)
f.close()
```

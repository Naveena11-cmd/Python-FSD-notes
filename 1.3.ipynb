**1. wap to convert given days into year, month and days**

```python
d=int(input("Enter number of days : "))
year=d//365
month=(d%365)//30
days=(d%365)%30
print(year," years", month," months", days," days")
```

**2. wap to count number of notes-10,20,50,100,200,500**

```python
amnt=int(input("Enter amount : "))
note500=amnt//500
note200=(amnt%500)//200
note100=((amnt%500)%200)//100
note50=(((amnt%500)%200)%100)//50
note20=((((amnt%500)%200)%100)%50)//20
note10=(((((amnt%500)%200)%100)%50)%20)%10
print(note500," notes of 500 ", note200," notes of 200 ",note100," notes of 100 ",note50, " notes of 50 ",note20, " notes of 20 ",note10," notes of 10")
```

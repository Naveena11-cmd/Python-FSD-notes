# 3.Sequence type : 
**list, tuple,range**

**1. List**
-to store multiple items in single variable
-ordered,**changable**,allows duplicate values
-list items are indexed
-square brackets are used

```python
l1=["apple", "banana", "cherry"]
print(l1)
print(type(l1))
print(l1[1])

l2=[1,6,7,9,3,2,3,6,7]
print(l2)
print(type(l2))

l3=["apple",3,3.5,"xyz"]
print(l3)
```

```python
l1=["apple", "banana", "cherry"]
print(l1)
print(type(l1))
print(l1[1])
l1[1]="strawberry"
print(l1)
```

**2. Tuple**: 
-to store multiple items in single variable
-ordered,**Unchangable**,allows duplicate values
-list items are indexed
-round brackets are used

```python
d=10,20,30
print(type(d))
```

```python
t1=("apple", "banana", "cherry")
print(t1)
print(type(t1))
print(t1[1])
t1[1]="strawberry"
print(t1)
```

**3. Range** :
It gives numbers from 0 to x-1
-range(5) : 5 is considered as end point and 0 is by default the start point. it will print numbers from 0 to 5
-range(1,5) : 1 is start point and 5 is end point, it will print 1,2,3,4
-range(1,10,2) : 1-start, 10-end, 2-step
It will print 1,3,5,7,9

```python
x=range(4)
print(x)
print(type(x))
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(-10,2):
    print(i)
```

# 4. Mapping Type

**Dict**
-stores data value in key:value pair
-ordered, changable and doesn't allow duplicates

```python
d={10:'lucky', 20:'arman',30:'dhairya'}
print(d)
print(type(d))
print(d[10])
d[10]="arman"
print(d)
```

# 6.Boolean Type

**bool**
-represents true or false

-any string is true except empty string

```python
print(bool("abc"))
print(bool(""))
print(bool(123))
print(bool([1,2,3,4]))
print(bool(0))
print(bool(" "))
```

# Global vs Local variable

```python
a="Python" #global 
def test() : 
    a="Java" #local
    print(a)
    
test()
print(a)
```

here a="Python" is global variable
a=Java is local varibale as it is declared inside a function

```python
a="Python"
def test() :
    global a
    a="Java"
    print(a)
    
test()
print(a)
```

if we declare a variable inside a function as 'global a' then it becomes a global variable

```python
a="Python"
def test() :
    global a
    a="Java"
    print(a)
    
print(a)
test()
print(a)
```

Before the function is called the globle variable is the one declared outside the function
and after calling the func, it will print the global variable declared inside the func
